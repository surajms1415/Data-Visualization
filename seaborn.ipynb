```python
s=sns.load_dataset("penguins")
s.head()
```

```python
#sns.barplot(x=s.island,y=s.bill_length_mm,hue="sex") 
sns.set(style="darkgrid")   # grid
sns.barplot(x="island",y="bill_length_mm",data=s,hue="sex",alpha=0.7)  # alpha - brightness
plt.show()
#hue_order=["Female","Male"] - to alter the bars 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var=np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True)
```

```python
var=np.linspace(1,10,10).reshape(2,5)
var
```

```python
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True)
```

```python
# to change the number
ar=np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"]])
ar
```

```python
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=ar,fmt="s")
```
